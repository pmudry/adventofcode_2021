Every step, the sea cucumbers in the east-facing herd attempt to move forward one location, then the sea cucumbers in the south-facing herd attempt to move forward one location. In this example, the sea cucumbers stop moving after 58 steps.

Find somewhere safe to land your submarine. What is the first step on which no sea cucumbers move?

````python
from scipy.ndimage import generic_filter
import numpy as np

# Get a more convenient representation for scipy
dataTypes = {".": 0, ">": 1, "v": 2}

def parse_input(fileName: str):
    with open(fileName) as f:
        lines = [[dataTypes[c] for c in list(x.strip())] for x in f]
    arr = np.array(lines)
    return arr

def move_east(a):
    if a[4] == 0 and a[3] == 1: return 1
    if a[4] == 1 and a[5] == 0: return 0
    return a[4]

def move_south(a):
    if a[4] == 0 and a[1] == 2: return 2
    if a[4] == 2 and a[7] == 0: return 0
    return a[4]

def compute(arr):
    # Stores all intermediary results for plotting later
    result_arr = []
    result_arr.append(np.copy(arr))
    i = 1
    while True:
        prev = arr
        arr = generic_filter(arr, move_east,  size=3, mode="wrap")
        arr = generic_filter(arr, move_south, size=3, mode="wrap")
        if np.array_equal(prev, arr):            
            return i, result_arr
            break
        result_arr.append(np.copy(arr))
        i += 1    

SAMPLE = parse_input('input_day25_sample.txt')
INPUT = parse_input('input_day25.txt')

steps, result_a = compute(SAMPLE)
assert (steps) == 58

print(f'Number of steps for sample input {steps}')

steps, result_a = compute(INPUT)
print(f'Number of steps for real input {steps}')
````

````python
import matplotlib.pyplot as plt
import matplotlib.animation as anim

data = result_a

def show_image(arr):
    plt.figure(figsize = (5,5))
    plt.tick_params(axis='both', colors='gray')
    plt.style.use('dark_background')
    plt.imshow(arr, vmin=abs(arr).min(), vmax=abs(arr).max(), origin='lower', cmap='magma', interpolation='antialiased')
    return fig

show_image(data[0])

fig = plt.figure(figsize = (5,5))
plt.tick_params(axis='both', colors='gray')
plt.style.use('dark_background')
im = plt.imshow(data[0], aspect='auto', vmin=abs(data).min(), vmax=abs(data).max(), cmap='magma', origin='lower')

def animate_func(i):
    im.set_array(data[i])
    return [fig]

move = anim.FuncAnimation(
        fig,
        animate_func,
        frames = len(data),
        interval = 40, blit=True
        )

move.save('day_25.mp4', fps=30)
````
