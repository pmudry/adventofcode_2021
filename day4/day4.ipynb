Bingo is played on a set of boards each consisting of a 5x5 grid of numbers. Numbers are chosen at random, and the chosen number is marked on all boards on which it appears. (Numbers may not appear on all boards.) If all numbers in any row or any column of a board are marked, that board wins. (Diagonals don't count.)

At this point, the third board wins because it has at least one complete row or column of marked numbers (in this case, the entire top row is marked: 14 21 17 24 4).

The score of the winning board can now be calculated. Start by finding the sum of all unmarked numbers on that board; in this case, the sum is 188. Then, multiply that sum by the number that was just called when the board won, 24, to get the final score, 188 * 24 = 4512.

To guarantee victory against the giant squid, figure out which board will win first. What will your final score be if you choose that board?

````python
from functools import reduce

DIM = 5

class Value:
    def __init__(self, value, present):
        self.value = value
        self.present = present
    
class Board:    
    def __init__(self, dim = DIM):
        self.dim = dim
        self.table = []
        self.latestValue = -1
        self.completionTurn = -1

    def checkPresent(self, value):
        for i in range(len(self.table)):
            for j in range(len(self.table[i])):
                elem = self.table[i][j]

                if(elem.value == value):
                    elem.present = True
                    return        

    def isComplete(self):
        # Check lines
        for i in range(len(self.table)):
            total = 0
            for j in range(len(self.table[i])):
                elem = self.table[i][j]

                if(elem.present):
                    total = total + 1

                if total == self.dim:
                    return True


        # Check columns
        for i in range(len(self.table)):
            total = 0
            for j in range(len(self.table[i])):
                elem = self.table[j][i]

                if(elem.present):
                    total = total + 1

                if total == self.dim:
                    return True

        return False                    

    def sumUnmarked(self):        
        # Flatten the list
        flattened = reduce(lambda x, y: x+y, self.table)
        # Add all the values if present
        return reduce(lambda x, y: x + (y.value if y.present == False else 0), flattened, 0)        

    # toString equivalent
    def __str__(self):
        r = 'Board\n'
        for i in range(len(self.table)):
            for j in range(len(self.table[i])):
                elem = self.table[i][j]
                r = r + f"{elem.value:02d}" + f"({'X' if elem.present else ' '})"
                if(j != len(self.table[i])-1):
                    r = r + '  '                
                    
            r = r + '\n'
        return r
````

````python
# Read and parse the file
f = open('input_day4.txt', 'r')

# Here are the drawn values, as a list of int
vals = f.readline()
drawn = [int(x) for x in vals.split(',')]

# Skip the empty line
f.readline()

# Create the list of boards

# Create a board from a list(list(int))
def createBoard(lines):
    assert len(lines) == DIM
    b = Board(DIM)

    for line in lines:
        values = line.split(" ")
        vals = []
        for v in values:
            vals.append(Value(int(v), False))

        b.table.append(vals)
    
    return b

boards = [] 

while True:    
    r = []    
    r.append(f.readline())

    # When reaching the end
    if r[0] == '':
        break

    for i in range(4):
        r.append(f.readline())    

    # Remove trailing \n and spaces as well as spaces in the middle
    for i in range(len(r)):        
        r[i] = " ".join(r[i].split())

    boards.append(createBoard(r))

    # Skip empty line at the end of each board
    f.readline()

# Show some extracted board for controlling
[print(board) for board in boards[0:1]]
f.close()
````
